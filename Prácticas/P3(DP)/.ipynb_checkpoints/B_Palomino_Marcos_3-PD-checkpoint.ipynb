<h3>El Héroe y la Princesa</h3>

<p>
Han capturado a la princesa del reino y nuestro valiente héroe debe rescatarla. La princesa ha sido encarcelada en la esquina inferior derecha de una fortaleza. La fortaleza está compuesta de <b>M x N</b> habitaciones cuadradas. Nuestro valiente héroe está inicialmente posicionado en la esquina superior izquierda de la fortaleza.
</p>
<p>
Nuestro héroe tiene una cantidad de vida representada por un número entero positivo. Si este número cae a 0, el héroe muere inmediatamente.
</p>
<p>
Algunas habitaciones están infestadas de monstruos, lo que hace que nuestro héroe pierda puntos de vida al pasar por ellas. Estas habitaciones las representamos con número negativos (cantidad de vida que perderá el héroe). Por otro lado, algunas habitaciones contienen corazones que reestablecerán la vida de nuestro héroe. Estas habitaciones las representamos con números positivos (cantidad de vida que se reestablecerá a nuestro héroe).
</p>
<p>
Para llegar a la princesa lo antes posible, nuestro héroe avanza únicamente hacia la derecha o hacia abajo a cada paso.
</p>

<img style="width:33%" src="img/hero.png">

<p>
Observad que las habitaciones con monstruos contienen un número negativo. Por otro lado, las habitaciones con corazones, contienen un número positivo. También vemos que tanto en la casilla superior izquierda (inicial) como en la inferior derecha (final) puede haber monstruos o corazones.
</p>
<p>
<i>Solución de la imagen de ejemplo: 3 puntos de vida.<br> Correspondiente al camino óptimo: Derecha -> Derecha -> Abajo -> Abajo</i>
</p>
<p>
    <b>¿Cuál será la cantidad mínima de salud inicial que deberá tener nuestro héroe para rescatar a la princesa?</b><br>
    Implementa una función que, dada una fortaleza, calcule el mínimo de vida inicial necesaria para completarla.
</p>

<div class="alert alert-danger" style="width:80%; margin:0 auto; padding">
<center><p><h3> Código </h3></p> </center>
<p>
<h3>INPUT</h3>
<ul>
<li><b>F</b>: Fortaleza representada como una lista de listas de números enteros.<br> E.g.: <i>[[-2,3,-1],[-3,-4,4],[-400,-1,-3]]</i></li>
</ul>
<br>
<h3>OUTPUT</h3>
<ul>
<li><b>HP</b>: Número entero correspondiente a la cantidad de vida mínima necesaria para rescatar a la princesa.</li>
</ul>
  
</p> 
</div>


````python
# Este ejercicio se puede resolver de diferentes maneras, una manera es utilizando recursion, aunque de esta manera nos 
# saldria una complejidad elevada, ya que tendriamos que calcular cosas que ya habíamos calculado.
# Con un algoritmo greedy, nos saldría mal, ya que en este caso, el resultado local no va a ser siempre el bueno.

# Mi propuesta para este ejercicio ha sido crear una matriz auxiliar donde iremos guardando en cada casilla el valor
# minimo para avanzar 

def hero(F):
    vidaHeroe = 0
    filas = len(F)
    columnas = len(F[0]) 

    # Creamos una matriz copia y inicializamos todos sus valores a 0
    sol = [[0 for i in range(columnas)] for j in range(filas)]  
    
    # Si el valor donde se encuentra la princesa es positivo, añadimos 1 a la casilla
    if F[filas-1][columnas-1] > 0:
        sol[filas-1][columnas-1] = 1
    else:
        # En caso que sea negativo, el valor minimo sera el absoluto de la casilla +1
        sol[filas-1][columnas-1] = abs(F[filas-1][columnas-1]) + 1
        
    # Primero completamos la ultima columna
    for x in range(filas-2,-1,-1):
        sol[x][columnas-1] = max(sol[x+1][columnas-1] - F[x][columnas-1], 1)
    
    # Despues la ultima fila
    for x in range(columnas-2,-1,-1):
        sol[filas-1][x] = max(sol[filas-1][x+1] - F[filas-1][x], 1)
        
    # Finalmente rellenamos toda la matriz empezando desde abajo 
    for i in range(filas-2,-1,-1):
        for j in range(columnas-2,-1,-1):
            sol[i][j] = max(min(sol[i + 1][j], sol[i][j + 1]) - F[i][j], 1)

    vidaHeroe = sol[0][0]
    return vidaHeroe
````

````python
# Pequeña tabla de test para diferentes matrices que nos dáis en el txt
F0 = [[-2,-3,3],[-5,-10,1],[10,30,-5]] # Para este caso, HP -> 7
F1 = [[-29, 82, 56, -73, -86, -79], [-72, 57, -30, 12, -30, -8], [55, 70, 43, -24, 14, 59], [29, 10, -79, 49, -92, -97], [-19, 54, -36, -64, -57, -69], [40, -5, -3, 33, 17, 52], [96, 97, 21, -65, -22, 77], [6, 18, -48, 64, 59, -89]] #30
F2 = [[26, 42, -66, -92, 26], [6, -21, -8, 19, -13], [85, 8, 51, 98, -16]] #1
F3 = [[-82, -62, -61], [-30, -41, 88]] #154

assert hero(F0) == 7, "resultado esperado 7"
assert hero(F1) == 30, "resultado esperado 30"
assert hero(F2) == 1, "resultado esperado 1"
assert hero(F3) == 154, "resultado esperado 154"

print("all test have been passed")
````

<h3>El extrañanemente resistente huevo</h3>

<p>
Tienes en tu poder $K$ huevos y acceso a un edificio de $F$ plantas. Sabemos que existe una planta $f$ tal que $0 \leq f \leq F$, en la que cualquier huevo lanzado en una planta superior a $f$ se romperá. Asimismo, cualquier huevo lanzado en la planta $f$ o inferior no se romperá.
</p>

<img src="img/resilient_egg.png">

<p>
    El objetivo es, lanzando huevos en distintas plantas, averiguar <b>con certeza absoluta</b> el valor de $f$. Es decir, a partir de qué planta comenzarán a romperse los huevos al caer.
    <br>
    <b>¿Cuál es el mínimo número de movimientos (dejar caer un huevo) necesarios para determinar $f$ con absoluta certeza?</b>
</p>
<p>
<ul>
<li>Todos los huevos son idénticos entre sí (misma resistencia).</li>
<li>Si un huevo resiste una caída desde el piso $y$, también resistirá para todos los pisos $x< y$.</li>
    <li>No puedes volver a lanzar un huevo que se haya roto</li>
</ul>
</p>

<p>
    Ejemplo: dado $K = 1$ huevo y $F = 2$ plantas.
    <ol>
        <li>Lanza el huevo desde el piso 1. Si se rompe, sabemos que $f = 0$.</li>
        <li>Si el huevo no se rompe, lánzalo desde el piso 2. Si se rompe, $f = 1$.</li>
        <li>Si no se rompe, $f = 2$</li>
    </ol>
    Por lo tanto, en el peor de los casos, nos ha hecho falta $2$ movimientos para determinar con certeza el valor de $f$.
</p>

<h4>Planteamiento</h4>
<p>
    Dado el caso $K = k$ y $F = n$. Probamos cuál es la planta óptima para lanzar el siguiente huevo.
    Esto es, probamos todas las plantas $1\leq x\leq n$.
    Si tiro un huevo en la planta $x$ (sea $D$ la matriz de programación dinámica):
    <ol>
        <li>Se rompe. Me quedan $k - 1$ huevos y sé que $f < x$. Subproblema: $D[k - 1][x - 1]$.</li>
        <li>No se rompe. Me quedan $k$ huevos y sé que $x \leq f \leq n$. Subproblema: $D[k][n - x]$.</li>
    </ol>
    Ya que debo saber con certeza el valor de $f$, debo quedarme el peor caso. Es decir, el caso con mayor movimientos requeridos. 
</p>
    $movimientos_x = max(D[k - 1][x - 1], D[k][n - x])$
<p>
    Esto me dice el número de movimientos necesarios después de lanzar el huevo desde el piso $x$. Dado que queremos encontrar el número óptimo de movimientos, calculamos los movimientos que necesitaría para todos los posibles valores de $x$ y me quedo con el mínimo. Finalmente, debemos sumar $1$ al resultado para contabilizar el intento que hacemos en el piso $x$ óptimo.
</p>
<p>
Resumiendo, la función recursiva para construir la matriz dinámica es:<br><br>
$D[k][n] = 1 + min(max(D[k - 1][x - 1], D[k][n - x])\quad \forall x\mid 1\leq x \leq n)$<br><br>
Donde:<br>
$\forall x\mid 1\leq x \leq n$ significa para todo $x$ en el intervalo $[1, n]$.
</p>

<div class="alert alert-danger" style="width:80%; margin:0 auto; padding">
<center><p><h3> Código </h3></p> </center>
<p>
<h3>INPUT</h3>
<ul>
<li><b>K</b>: número de huevos.</li>
<li><b>F</b>: número de plantas.</li>
</ul>
<br>
<h3>OUTPUT</h3>
<ul>
<li><b>M</b>: Número mínimo de movimientos necesarios para conocer con certeza el valor $f$.</li>
</ul>
  
</p> 
</div>


````python
def eggs(K, F): 
    # Creamos una matriz[i][j] donde en cada casilla pondremos el minimo de movimientos necesarios para i huevos y j plantas
    # la matriz empieza contando desde 0
    
    # En el caso que tengamos 0 huevos o 0 plantas, el numero de tiradas serán 0
    sol = [[0 for i in range(F+1)] for j in range(K+1)]  
    
    # Lista donde añadiré el maximo de una tirada de huevo en caso de que se rompa o no, desde un numero de plantas
    listaPeoresCasos = []
    
    # En el caso que tengamos 1 planta, tendremos que tirar unicamente 1 vez el huevo
    for x in range(1,K+1):
        sol[x][1] = 1
        
    # En caso que tengamos 1 huevo, los minimos movimientos(en el peor caso) serían el numero de plantas
    for y in range(1, F+1):
        sol[1][y] = y
        
    # Completamos la matriz dinamica
    for i in range(2,K+1):
        for j in range(2,F+1):
            # Iteramos sobre las plantas donde haremos las tiradas de los subproblemas
            for x in range(1,j+1):
                
                # Añadimos el peor caso, entre si se rompe el huevo o no desde el subproblema con "i" huevos y "x" plantas
                listaPeoresCasos.append(1 + max(sol[i-1][x-1],sol[i][j-x]))
            
            # De todos los peores casos, cojeremos el mínimo 
            sol[i][j] = min(listaPeoresCasos)
            
            # Ponemos la lista de nuevo vacia para la siguiente iteración
            listaPeoresCasos = []

    # El valor se encontrará en la ultima casilla
    return sol[K][F]
````

````python
# Pequeña tabla de test para probar algunos de los ejemplos que nos dáis en el txt
assert eggs(6,74) == 7
assert eggs(49,33) == 6
assert eggs(25,54) == 6
assert eggs(39,11) == 4
assert eggs(2,6) == 3
assert eggs(3,14) == 4

print("all test have been passed")
````
